```python
import io
from bs4 import BeautifulSoup
import urllib2
```

```python
def writeToFile(table,column):
    """write a bs4 table to a file. Each table is passed as a list.
    [table name (e.g. URL), bs4 table]
    the file is named to relate to the URL"""
    fileName = table[0] + '.txt'
    f = io.open(fileName, 'w', encoding='utf8')
    text = []
    for trs in table[1].findAll("tr")[1:]: #ignore the first row (e.g. headings)
        tds = trs.findAll("td")
        text.append(tds[column].text)
        #t = tds[column].text
        #print t
        #print "----------"

    #print text
    for item in text:
      f.write("%s\n" % item)

    f.close()
```

```python
def pulltable(wiki):
    header = {'User-Agent': 'Mozilla/5.0'} #Needed to prevent 403 error on Wikipedia
    req = urllib2.Request(wiki,headers=header)
    page = urllib2.urlopen(req)
    soup = BeautifulSoup(page,"html.parser")

    table = soup.find("table", { "class" : "sortable" })
    name = wiki.split("/")[4]
    return [name,table]
#print table
```

```python
urls = [
    "https://en.wikipedia.org/wiki/List_of_designated_terrorist_groups"    
]
```

```python
for url in urls:
    print url.split("/")[4]
```

```python
tables = []
for url in urls:
    #print pulltable(url)
    tables.append(pulltable(url))
print tables
```

```python
for table in tables:
    writeToFile(table,1)
```
